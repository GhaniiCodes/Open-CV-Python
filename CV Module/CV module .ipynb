```python
cam = cv.VideoCapture(0)

while True:
    
    ret, frame = cam.read()

    if not ret:
        print("Cam Not Found")
        break
    
    # Resize the frame to half of the original size
    Resized_Frame = cv.resize(frame, (0, 0), fx=0.5, fy=0.5)
    
    # Rotate the resized frame
    Rotation_180 = cv.rotate(Resized_Frame, cv.ROTATE_180)
    
    # Rotate 90 degrees clockwise and resize again
    Rotation_90_ClockWise = cv.rotate(Resized_Frame, cv.ROTATE_90_CLOCKWISE)
    Rotation_90_ClockWise_Resized = cv.resize(Rotation_90_ClockWise, (0, 0), fx=0.5, fy=0.5)
    Rotation_90_ClockWise_Resized_ColorChanged = cv.cvtColor(Rotation_90_ClockWise_Resized, cv.COLOR_HLS2BGR)  # Convert to BGR
    
    # Rotate 90 degrees counterclockwise and resize again
    Rotation_90_CounterClockWise = cv.rotate(Resized_Frame, cv.ROTATE_90_COUNTERCLOCKWISE)
    Rotation_90_CounterClockWise_Resized = cv.resize(Rotation_90_CounterClockWise, (0, 0), fx=0.5, fy=0.5)
    Rotation_90_CounterClockWise_Resized_ColorChanged = cv.cvtColor(Rotation_90_CounterClockWise_Resized, cv.COLOR_LAB2BGR)  # Convert to BGR
    
    # Ensure all frames have the same height and width by resizing to match Resized_Frame's dimensions
    Rotation_180 = cv.resize(Rotation_180, Resized_Frame.shape[1::-1])
    Rotation_90_ClockWise_Resized = cv.resize(Rotation_90_ClockWise_Resized, Resized_Frame.shape[1::-1])
    Rotation_90_ClockWise_Resized_ColorChanged = cv.resize(Rotation_90_ClockWise_Resized_ColorChanged, Resized_Frame.shape[1::-1])
    Rotation_90_CounterClockWise_Resized = cv.resize(Rotation_90_CounterClockWise_Resized, Resized_Frame.shape[1::-1])
    Rotation_90_CounterClockWise_Resized_ColorChanged = cv.resize(Rotation_90_CounterClockWise_Resized_ColorChanged, Resized_Frame.shape[1::-1])

    # Horizontal concatenations
    Combined_frame_1 = cv.hconcat([Resized_Frame, Rotation_180, Rotation_90_ClockWise_Resized_ColorChanged])
    Combined_frame_2 = cv.hconcat([Rotation_90_CounterClockWise_Resized, Rotation_90_CounterClockWise_Resized_ColorChanged, Rotation_90_ClockWise_Resized])

    # Vertical concatenation
    Union_Frame = cv.vconcat([Combined_frame_1, Combined_frame_2])
    
    # Display the final frame
    cv.imshow("WebCam", Union_Frame)
    
    # Break the loop on pressing 'a'
    if cv.waitKey(1) == ord("a"):
        break

# Release resources
cam.release()
cv.destroyAllWindows()
```

Color Detection 

```python
cam = cv.VideoCapture(0)

while True:
    ret , frame = cam.read()
    
    #Converting To HSV
    HSV = cv.cvtColor(frame , cv.COLOR_BGR2HSV)
    
    #Giving a Range of HSV Values 
    mask = cv.inRange(HSV, (100,50,50), (130,255,255))  
    
    #Showing Result For the Given Range 
    result = cv.bitwise_and(frame , frame , mask=mask)
    
    cv.imshow("Cam" , result)
    
    if ret == False:
        print ("Camera Not Found")
        break
    if cv.waitKey(1) == ord("a"):
        break
    
cam.release()
cv.destroyAllWindows()
```

Detection and Tracking of a Color 

```python
# Lower bound HSV
h_lower = int(input("Enter lower H value (0-179): "))
s_lower = int(input("Enter lower S value (0-255): "))
v_lower = int(input("Enter lower V value (0-255): "))

# Upper bound HSV
h_upper = int(input("Enter upper H value (0-179): "))
s_upper = int(input("Enter upper S value (0-255): "))
v_upper = int(input("Enter upper V value (0-255): "))

cam = cv.VideoCapture(0)

while True:
    ret, frame = cam.read()
    
    if ret == False:
        print("Camera Not Found")
        break

    # Convert frame to HSV
    hsv_Frame = cv.cvtColor(frame, cv.COLOR_BGR2HSV)
    
    # Apply the HSV range to create a mask
    mask = cv.inRange(hsv_Frame, (h_lower, s_lower, v_lower), (h_upper, s_upper, v_upper))  
    
    
    Result = cv.bitwise_and(frame , frame , mask=mask)
    
    # Find the non-zero points in the mask
    points = cv.findNonZero(mask)

    if points is not None:
        # Compute the bounding box for the non-zero points
        x, y, w, h = cv.boundingRect(points)
        
        # Draw a rectangle around the detected area
        cv.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

    # Display the result
    cv.imshow("Cam", frame)
    cv.imshow("Frame" , Result)
    
    # Break the loop when 'a' is pressed
    if cv.waitKey(1) == ord("a"):
        break

# Release the camera and close the window
cam.release()
cv.destroyAllWindows()
```

Face And Eye Detection 

```python
# Initialize the camera
cam = cv.VideoCapture(0)

# Load Haar cascades for face and eye detection
face_cascade = cv.CascadeClassifier(cv.data.haarcascades + 'haarcascade_frontalface_default.xml')
eye_cascade = cv.CascadeClassifier(cv.data.haarcascades + 'haarcascade_eye.xml')

while True:
    # Capture frame-by-frame
    ret, Frame = cam.read()
    
    if ret == False:
        print("Camera Not found")
        break
    
    # Convert the frame to grayscale
    Gray_Frame = cv.cvtColor(Frame, cv.COLOR_BGR2GRAY)
    
    # Detect faces in the frame
    faces = face_cascade.detectMultiScale(Gray_Frame, 1.05, 5)
    
    # Draw rectangle around detected faces
    for (x, y, w, h) in faces:
        cv.rectangle(Frame, (x, y), (x + w, y + h), (255, 125, 0), 2)
        roi_Gray = Gray_Frame[y:y+h, x:x+w]
        roi_Color = Frame[y:y+h, x:x+w]
        
        # Detect eyes in the face region
        eyes = eye_cascade.detectMultiScale(roi_Gray, 1.05, 5)
        for (ex, ey, ew, eh) in eyes:
            # Corrected rectangle for eyes (x, y, width, height)
            cv.rectangle(roi_Color, (ex, ey), (ex + ew, ey + eh), (0, 125, 90), 2)
    
    # Display the resulting frame
    cv.imshow("Web_camera", Frame)
    
    # Break the loop if the 'a' key is pressed
    if cv.waitKey(1) == ord('a'):
        break

# Release the camera and close all windows
cam.release()
cv.destroyAllWindows()
```

Motion Detection  

```python
cap = cv.VideoCapture(r"D:\PYTHON\CV Module\Road.mp4")

ret, frame1 = cap.read()
frame1_gray = cv.cvtColor(frame1, cv.COLOR_BGR2GRAY)
frame1_gray = cv.GaussianBlur(frame1_gray, (21, 21), 0)

while cap.isOpened():

    ret, frame2 = cap.read()
    
    if ret == False:
        print ("Video Not Found")
        break
    
    # Convert the frame to grayscale and apply GaussianBlur
    frame2_gray = cv.cvtColor(frame2, cv.COLOR_BGR2GRAY)
    frame2_gray = cv.GaussianBlur(frame2_gray, (21, 21), 0)
    
    # Compute the absolute difference between the current frame and previous frame
    frame_Difference = cv.absdiff(frame1_gray, frame2_gray)
    
    # Apply a binary threshold to the difference image
    _, thresh_frame = cv.threshold(frame_Difference, 25, 255, cv.THRESH_BINARY)
    
    # Dilate the threshold image to fill in holes, making it easier to detect contours
    thresh_frame = cv.dilate(thresh_frame, None, iterations=2)
    
    # Find contours of the threshold image
    contours, _ = cv.findContours(thresh_frame.copy(), cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
    
    # Draw rectangles around the contours (detected moving objects)
    for contour in contours:
        
            (x, y, w, h) = cv.boundingRect(contour)
            cv.rectangle(frame2, (x, y), (x+w, y+h), (0, 255, 0), 2)
    
    # Display the resulting frames
    cv.imshow("Motion Detection", frame2)
    
    
    # Update the previous frame to be the current frame
    frame1_gray = frame2_gray
    
    # Break the loop if 'q' is pressed
    if cv.waitKey(10) == ord('a'):
        break

# Release the capture object and close all windows
cap.release()
cv.destroyAllWindows()
```
